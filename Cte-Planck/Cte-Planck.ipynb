```python
def piecewise(data,sep, int_x1, int_x2, x,y, ylim, file):
    x1,x2 = data['V'][:sep].to_numpy(),data['V'][sep:].to_numpy()
    y1,y2 = data['I'][:sep].to_numpy(),data['I'][sep:].to_numpy()
    
    reg1 = LinearRegression().fit(x1.reshape(-1,1), y1)
    reg2 = LinearRegression().fit(x2.reshape(-1,1), y2)
    
    x1val = np.linspace(int_x1[0],int_x1[1],10)
    x2val = np.linspace(int_x2[0],int_x2[1],10)
    
    fig,axs = plt.subplots(figsize = (10,6))
    
    color = data['cor'].unique()[0]
    
    axs.plot(data['V'],data['I'], 'D', markersize = 5, color = color, label = 'LI ' + str(color)+ ' LED data',)
    axs.plot(x1val, reg1.coef_*x1val + reg1.intercept_, '--', color = 'blue',label = 'Piecewise')
    axs.plot(x2val, reg2.coef_*x2val + reg2.intercept_, '--', color = 'blue',)
    axs.vlines(x,ylim,y,linestyle = '--',label = 'Vmin')
    
    plt.ylim(ylim[0],ylim[1])
    axs.set_xlabel(r'$V(V)$', fontsize = 16)
    axs.set_ylabel(r'$I(mA)$', fontsize = 16)
    
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    
    axs.legend(fontsize = 15)
    plt.savefig(file)
    plt.show()
```

```python
def plot(df):
    fig,axs = plt.subplots(figsize = (10,8))
    
    axs.plot(df['V'],df['I'], 'o', color = df['cor'].unique()[0], markersize = 6, linewidth = 3)
    
    axs.set_xlabel('V', fontsize = 15)
    axs.set_ylabel('I', fontsize = 15)
    
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    
    axs.legend(fontsize = 15)
    
    plt.show(True)
```
