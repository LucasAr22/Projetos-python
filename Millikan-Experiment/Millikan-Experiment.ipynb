```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from scipy.constants import e
```

### Importando e organizando os dados

```python
df = pd.read_csv("Cargas3.txt",dtype = float)
df.columns = ['index', 'Q1', 'Q2']

df['Q2'] = np.abs(df['Q2'])
df['Q1'] = np.abs(df['Q1'])

data = df[df['Q1'] != 0].copy()
data = data[data['Q2'] != 0].copy()
#df['Q1'] = 3.3356*10**(-10)*np.abs(df['Q1'])
#df['Q2'] = 3.3356*10**(-10)*np.abs(df['Q2'])
```

### Dados $<Q>$

#### Histograma

```python
fig,axs = plt.subplots(figsize = (10,6))
axs.hist((data['Q1']+data['Q2'])/2, color = 'midnightblue', label='Data', bins = 35, align = 'mid')


axs.set_ylabel('Ocorrências', fontsize = 20)
axs.set_xlabel('$<Q>$', fontsize = 20)

plt.xlim(0,1.8*10**(-17))
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

#### Carga Elétron

```python
y = (data['Q1'] + data['Q2'])/2
x = y/min(y)

x = x.to_numpy()
reg = LinearRegression().fit(x.reshape(-1,1), y)
fig,axs = plt.subplots(figsize = (10,6))

axs.plot(x,y,'D',markersize = 7,label='Dados', color = 'midnightblue')
axs.plot(x,x*reg.coef_,'-',label='Fit', color = 'black',linewidth = 1)
axs.set_xlabel(r'$n$', fontsize = 20)
axs.set_ylabel(r'$<Q>$', fontsize = 20)

axs.legend(fontsize = 15)


plt.xlim(0,90)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

print('e = ',reg.coef_[0])
```
